```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import json
```

```python
file = open("fbooks.json")
data = json.load(file)
```

```python
#print("abcd")
for genre in data:
    booklist = data[genre]
    newbooklist = []
    for book in booklist:
        bookname = list(book.keys())[0]
        url = "http://libgen.lc/search.php?req=" + ('+'.join(bookname.split()))
        print("Old bookname :" , bookname)
        try:
            r = requests.get(url)
            soup = BeautifulSoup(r.content)
            result = soup.find('tr' , attrs={'bgcolor' : '#C6DEFF'})
            columns = result.findAll('td')
            id = columns[0].text
            authors = [author.text for author in columns[1].findAll('a')]
            publisher = columns[3].text
            year = columns[4].text
            pages = columns[5].text
            language = columns[6].text
            size = columns[7].text
            download = columns[9].find('a')['href']
            bookurl = 'http://libgen.lc/' + columns[2].find('a' , attrs={'id' : id})['href']
            print('Book Url :' , bookurl)
            temp = columns[2].find('a' , attrs={'id' : id})
            try:
                temp.font.decompose()
                temp.i.decompose()
            except:
                pass
            bookname = temp.text
            # print(authors , publisher , year , pages , language , size , download , bookurl)

            r = requests.get(bookurl)
            soup = BeautifulSoup(r.content)
            image = 'http://libgen.lc/' + soup.find('img' , attrs={'style':'padding: 5px'})['src']
        
            # print(image)
            # print(result)

            book['authors'] = authors
            book['image'] = image
            book['download'] = download
            book['publisher'] = publisher
            book['year'] = year
            book['pages'] = pages
            book['language'] = language 
            book['size'] = size

            #print(book)
            print("New bookname :" , bookname)
            print()
            newbooklist.append({bookname : book})
        except:
            print("Err")
        #break
    data[genre] = newbooklist
    #break
```

```python
fdata = {}
for genre in data:
    print(genre)
    fdata[genre] = []
    print()
    for book in data[genre]:
        fdict = {}
        bookname = list(book.keys())[0]
        inbook = book[bookname]
        fdict['authors'] = inbook['authors']
        fdict['image'] = inbook['image']
        fdict['download'] = inbook['download']
        fdict['publisher'] = inbook['publisher']
        fdict['year'] = inbook['year']
        fdict['pages'] = inbook['pages']
        fdict['language'] = inbook['language']
        fdict['size'] = inbook['size']

        bookdata = fdict
        fdata[genre].append({bookname : bookdata})
```

```python
for genre in fdata:
    print(genre)
    for book in fdata[genre]:
        bookname = list(book.keys())[0]
        for attr in book[bookname]:
            print(attr , ':' , book[bookname][attr])
```
